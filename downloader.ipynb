```python
import requests
from bs4 import BeautifulSoup as bs
import time
```

```python
section_urls = [
"https://f-droid.org/en/categories/development/",
"https://f-droid.org/en/categories/games/",
"https://f-droid.org/en/categories/graphics/",
"https://f-droid.org/en/categories/internet/",
"https://f-droid.org/en/categories/money/",
"https://f-droid.org/en/categories/multimedia/",
"https://f-droid.org/en/categories/navigation/",
"https://f-droid.org/en/categories/phone-sms/",
"https://f-droid.org/en/categories/reading/",
"https://f-droid.org/en/categories/science-education/",
"https://f-droid.org/en/categories/security/",
"https://f-droid.org/en/categories/sports-health/",
"https://f-droid.org/en/categories/system/",
"https://f-droid.org/en/categories/theming/",
"https://f-droid.org/en/categories/time/",
"https://f-droid.org/en/categories/writing/",
]
```

```python
apk_index_urls = []
SLEEP_TIME = 2
```

```python
for url in section_urls:
    response = requests.get(url)
    soup = bs(response.content, "html.parser")
    page_num = 2
    while soup.find('h2').text != '404 Page Not Found':
        for h3 in soup.find_all("h3"):
            for a in h3.find_all("a"):
                apk_index_urls.append('https://f-droid.org' + a.get("href"))
                print('https://f-droid.org' + a.get("href"))
        time.sleep(SLEEP_TIME)
        response = requests.get(url + str(page_num))
        soup = bs(response.content, "html.parser")
        page_num += 1
    print(soup.find('h2').text)
    print(url + " Done")
    print("---------------------------------")
    time.sleep(SLEEP_TIME)
```

```python
apk_download_urls = []
with open('apk_index_urls.txt', 'r') as f:
    apk_index_urls = f.read().split('\n')
print(apk_index_urls)
```

```python
for url in apk_index_urls:
    response = requests.get(url)
    soup = bs(response.text, "html.parser")
    download_url = soup.find('p', {'class': 'package-version-download'}).find('a')['href']
    apk_download_urls.append(download_url)
    print(download_url)
    time.sleep(SLEEP_TIME)
print('Done')
print('------------------------------------------')
```

```python
for url in apk_download_urls:
    with open("apk_download_urls.txt", "a") as f:
        f.write(url + "\n")

print("Done")
```
